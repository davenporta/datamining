```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
class LinearRegression:
    """Linear regression with numpy arrays"""
    def __init__ (self, A, b):
        self.A = A
        self.b = b

    def fit(self):
        """returns tuple (m,b)"""
        self.sigma = np.linalg.inv(self.A.T.dot(self.A)).dot(self.A.T.dot(self.b))
        return (self.sigma[0], self.sigma[1])
```

```python
def noise(array):
    for i in range(len(array)):
        n = np.random.normal(0,0.6)
        array[i] += n
    return array
```

```python
testcase = np.array([1,2,3,4,5,6], dtype=float)
y = noise(testcase)
x = np.arange(1,len(y)+1)
A = np.array([x,np.ones(x.shape)]).T

m,b = LinearRegression(A,y).fit()

# best fit line with coordinates (b,m)
line = (b,m)
points_on_line = np.array([[x_value,line[0] + x_value * line[1]] for x_value in np.arange(1,len(y)+1)])

# set up the plot
fig = plt.scatter(x,y, s=100)
plt.axis([0, len(y)+1, 0, 7])

# add the line
plt.plot([0, len(y)+1], [line[0], line[0] + (len(y)+1)* line[1]], color='g', linestyle='-', linewidth=2)

# point out the error terms
for (r,s),t in zip(points_on_line,testcase):
    plt.plot([r,r],[t,s],color='k',linestyle=':',linewidth=1)

plt.show()
```
