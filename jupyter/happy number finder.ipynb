```python
n = 10000
allopt = list(range(2,n+1))
happy = [1]
sad = []
count = 0
while len(allopt) != 0:
    count += 1
    num = allopt[0]
    buffer_list = []
    while num not in buffer_list:
        buffer_list.append(num)
        if num in happy:
            num = 1
            pass
        num = sum([int(i)**2 for i in list(str(num))])
    if 1 in buffer_list:
        happy += buffer_list
    else:
        sad += buffer_list
    allopt = [x for x in allopt if x not in buffer_list]

happy = set(happy)
sad = set(sad)
```
